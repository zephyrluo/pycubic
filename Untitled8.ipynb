```c
#include <time.h>
#include <stdbool.h>
#include <string.h>
#include <stdio.h>
#include <stdlib.h>

/* ==================== 常量 ==================== */
#define GS              18
#define NUM_OBSTACLES   10
#define MAX_BODY        (GS*GS)
#define MAX_HEAP        (GS*GS*4)
#define BFS_QUEUE_SIZE  (GS*GS)
#define MAX_ASTAR_PATH  (GS*GS)

static const int DY[4] = {-1, 0, 1, 0};
static const int DX[4] = { 0, 1, 0,-1};

const char DIR_CHARS[4] = {'W', 'D', 'S', 'A'};

char raw_map[20][25];

/* ==================== 蛇身 ==================== */
typedef struct { int y, x; } Pos;

typedef struct {
    Pos buf[MAX_BODY];
    int head, tail, len;
} BodyList;

static void body_init(BodyList *b, Pos p) {
    b->buf[0] = p; b->head = 0; b->tail = 1; b->len = 1;
}
static inline Pos body_get(const BodyList *b, int idx) {
    return b->buf[(b->head + idx) % MAX_BODY];
}
static inline Pos body_head(const BodyList *b) { return b->buf[b->head]; }
static inline Pos body_tail(BodyList *b) {
    return b->buf[(b->head + b->len - 1) % MAX_BODY];
}
static inline bool body_contains(BodyList *b, Pos p) {
    for (int i = 0; i < b->len; i++) {
        Pos bp = body_get(b, i);
        if (bp.y == p.y && bp.x == p.x) return true;
    }
    return false;
}
static inline void body_push_front(BodyList *b, Pos p) {
    b->head = (b->head - 1 + MAX_BODY) % MAX_BODY;
    b->buf[b->head] = p; b->len++;
}
static inline void body_add_tail(BodyList *b, Pos p) {
    b->buf[b->tail] = p;
    b->tail = (b->tail + 1 + MAX_BODY) % MAX_BODY;
    b->len++;
}
static inline Pos body_pop_back(BodyList *b) {
    b->tail = (b->tail - 1 + MAX_BODY) % MAX_BODY;
    b->len--;
    return b->buf[b->tail];
}
static inline void body_copy(BodyList *dst, const BodyList *src) { *dst = *src; }

/* ==================== 全局 body 位图 ==================== */
static bool g_body_grid[GS][GS];

static inline void body_grid_build(const BodyList *b) {
    memset(g_body_grid, 0, sizeof(g_body_grid));
    for (int i = 0; i < b->len; i++) {
        Pos p = body_get(b, i);
        g_body_grid[p.y][p.x] = true;
    }
}

/* ==================== 环境 ==================== */
typedef struct {
    int grid_size, n_steps_grow, num_obstacles;
    BodyList body;
    int direction, steps, growth_buffer, score;
    bool obstacles[GS][GS];
    Pos food;
} SnakeEnv;

static unsigned long long rng_state = 0xdead;
static void rng_seed(unsigned long long s) { rng_state = s ? s : 1; }
static unsigned long long rng_next(void) {
    rng_state ^= rng_state << 13;
    rng_state ^= rng_state >> 7;
    rng_state ^= rng_state << 17;
    return rng_state;
}
static int rng_randint(int lo, int hi) {
    return lo + (int)(rng_next() % (unsigned long long)(hi - lo));
}

static void env_spawn_food(SnakeEnv *e) {
    while (1) {
        int y = rng_randint(0, e->grid_size);
        int x = rng_randint(0, e->grid_size);
        Pos p = {y, x};
        if (!body_contains(&e->body, p) && !e->obstacles[y][x]) {
            e->food = p; return;
        }
    }
}

static void env_reset(SnakeEnv *e) {
    int center = e->grid_size / 2;
    body_init(&e->body, (Pos){center, center});
    body_grid_build(&e->body);
    e->direction = 0; e->steps = 0; e->growth_buffer = 0; e->score = 0;
    memset(e->obstacles, 0, sizeof(e->obstacles));
    int placed = 0;
    while (placed < e->num_obstacles) {
        int y = rng_randint(0, e->grid_size);
        int x = rng_randint(0, e->grid_size);
        if (y == center && x == center) continue;
        if (center-2<=y&&y<=center+2&&center-2<=x&&x<=center+2) continue;
        if (!e->obstacles[y][x]) { e->obstacles[y][x] = true; placed++; }
    }
    env_spawn_food(e);
}

static int env_step(SnakeEnv *e, int action_dir) {
    e->steps++; e->direction = action_dir;
    Pos head = body_head(&e->body);
    Pos nh = {head.y+DY[action_dir], head.x+DX[action_dir]};
    if (nh.y<0||nh.y>=e->grid_size||nh.x<0||nh.x>=e->grid_size) return 1;
    if (e->obstacles[nh.y][nh.x]) return 2;
    bool eat_food = (nh.y==e->food.y&&nh.x==e->food.x);
    bool step_grow = (e->steps%e->n_steps_grow==0);
    if (eat_food||step_grow) e->growth_buffer++;
    if (eat_food) { e->score+=10; env_spawn_food(e); }
    if (e->growth_buffer>0) e->growth_buffer--;
    else body_pop_back(&e->body);
    if (body_contains(&e->body,nh)) return 3;
    body_push_front(&e->body,nh);
    return 0;
}

/* ==================== AI ==================== */
typedef struct {
    int grid_size;
    bool init_obstacles[GS][GS];
    Pos  hamilton_cycle[GS*GS];
    int  hamilton_len;
    int  hamilton_pos[GS][GS];
    bool hamilton_valid;
} SnakeAI;

/* ============================================================
 * 哈密顿回路构建（三阶段）
 * ============================================================ */
static int  g_dfs_path[GS*GS];
static bool g_dfs_visited[GS][GS];
static int  g_dfs_total;
static bool g_dfs_found;
static long g_dfs_calls;
#define DFS_MAX_CALLS 2000000L

static int warnsdorff_score(int ny, int nx, bool obstacles[GS][GS], int gs) {
    int cnt = 0;
    for (int d = 0; d < 4; d++) {
        int y2=ny+DY[d], x2=nx+DX[d];
        if (y2>=0&&y2<gs&&x2>=0&&x2<gs&&!obstacles[y2][x2]&&!g_dfs_visited[y2][x2])
            cnt++;
    }
    return cnt;
}

static void hamilton_dfs(int y, int x, int depth, bool obstacles[GS][GS], int gs) {
    if (g_dfs_found||g_dfs_calls>DFS_MAX_CALLS) return;
    g_dfs_calls++;
    g_dfs_path[depth] = y*GS+x;

    if (depth==g_dfs_total-1) {
        int sy=g_dfs_path[0]/GS, sx=g_dfs_path[0]%GS;
        if (abs(y-sy)+abs(x-sx)==1) g_dfs_found=true;
        return;
    }

    int nbr_d[4], nbr_score[4], nbr_cnt=0;
    for (int d=0;d<4;d++) {
        int ny=y+DY[d],nx=x+DX[d];
        if (ny<0||ny>=gs||nx<0||nx>=gs||obstacles[ny][nx]||g_dfs_visited[ny][nx]) continue;
        int sc=warnsdorff_score(ny,nx,obstacles,gs);
        int pos=nbr_cnt++; nbr_d[pos]=d; nbr_score[pos]=sc;
        for (int k=pos;k>0&&nbr_score[k]<nbr_score[k-1];k--) {
            int td=nbr_d[k]; nbr_d[k]=nbr_d[k-1]; nbr_d[k-1]=td;
            int ts=nbr_score[k]; nbr_score[k]=nbr_score[k-1]; nbr_score[k-1]=ts;
        }
    }
    for (int i=0;i<nbr_cnt&&!g_dfs_found;i++) {
        int d=nbr_d[i], ny=y+DY[d], nx=x+DX[d];
        g_dfs_visited[ny][nx]=true;
        hamilton_dfs(ny,nx,depth+1,obstacles,gs);
        if (!g_dfs_found) g_dfs_visited[ny][nx]=false;
    }
}

static void build_bfs_order(SnakeAI *ai, bool obstacles[GS][GS], int gs) {
    bool visited[GS][GS]; memset(visited,0,sizeof(visited));
    Pos queue[GS*GS]; int qh=0,qt=0;
    int sy=-1,sx=-1;
    for (int y=0;y<gs&&sy<0;y++)
        for (int x=0;x<gs&&sy<0;x++)
            if (!obstacles[y][x]) { sy=y; sx=x; }
    if (sy<0) return;
    visited[sy][sx]=true; queue[qt++]=(Pos){sy,sx};
    ai->hamilton_len=0;
    while (qh<qt) {
        Pos cur=queue[qh++];
        int idx=ai->hamilton_len;
        ai->hamilton_cycle[idx]=cur;
        ai->hamilton_pos[cur.y][cur.x]=idx;
        ai->hamilton_len++;
        for (int d=0;d<4;d++) {
            int ny=cur.y+DY[d],nx=cur.x+DX[d];
            if (ny>=0&&ny<gs&&nx>=0&&nx<gs&&!obstacles[ny][nx]&&!visited[ny][nx]) {
                visited[ny][nx]=true; queue[qt++]=(Pos){ny,nx};
            }
        }
    }
    ai->hamilton_valid=false;
}

static bool verify_hamilton_path(SnakeAI *ai) {
    int len=ai->hamilton_len;
    if (len<=1) return false;
    for (int i=0;i<len-1;i++) {
        Pos a=ai->hamilton_cycle[i],b=ai->hamilton_cycle[i+1];
        if (abs(a.y-b.y)+abs(a.x-b.x)!=1) return false;
    }
    Pos first=ai->hamilton_cycle[0],last=ai->hamilton_cycle[len-1];
    return (abs(first.y-last.y)+abs(first.x-last.x)==1);
}

static bool try_dfs_from(SnakeAI *ai, int sy, int sx,
                          bool obstacles[GS][GS], int gs, int free_cells) {
    memset(g_dfs_visited,0,sizeof(g_dfs_visited));
    g_dfs_total=free_cells; g_dfs_found=false; g_dfs_calls=0;
    g_dfs_visited[sy][sx]=true;
    hamilton_dfs(sy,sx,0,obstacles,gs);
    if (!g_dfs_found) return false;
    ai->hamilton_len=free_cells;
    for (int i=0;i<free_cells;i++) {
        int y=g_dfs_path[i]/GS, x=g_dfs_path[i]%GS;
        ai->hamilton_cycle[i]=(Pos){y,x};
        ai->hamilton_pos[y][x]=i;
    }
    if (verify_hamilton_path(ai)) { ai->hamilton_valid=true; return true; }
    memset(ai->hamilton_pos,-1,sizeof(ai->hamilton_pos));
    ai->hamilton_len=0;
    return false;
}

static void ai_init(SnakeAI *ai, int grid_size, bool obstacles[GS][GS]) {
    ai->grid_size=grid_size;
    memcpy(ai->init_obstacles,obstacles,sizeof(ai->init_obstacles));
    memset(ai->hamilton_pos,-1,sizeof(ai->hamilton_pos));
    ai->hamilton_len=0; ai->hamilton_valid=false;

    int free_cells=0, start_y=-1, start_x=-1;
    for (int y=0;y<grid_size;y++)
        for (int x=0;x<grid_size;x++)
            if (!obstacles[y][x]) {
                free_cells++;
                if (start_y<0) { start_y=y; start_x=x; }
            }
    if (free_cells==0) return;

    /* 阶段1：从默认起点尝试 */
    if (try_dfs_from(ai,start_y,start_x,obstacles,grid_size,free_cells)) return;

    /* 阶段2：从四个角落最近的非障碍格重试 */
    int corners[4][2]={{0,0},{0,grid_size-1},{grid_size-1,0},{grid_size-1,grid_size-1}};
    for (int ci=0;ci<4&&!ai->hamilton_valid;ci++) {
        int cy=corners[ci][0],cx=corners[ci][1];
        int best_y=-1,best_x=-1,best_dist=9999;
        for (int y=0;y<grid_size;y++)
            for (int x=0;x<grid_size;x++)
                if (!obstacles[y][x]) {
                    int dist=abs(y-cy)+abs(x-cx);
                    if (dist<best_dist) { best_dist=dist; best_y=y; best_x=x; }
                }
        if (best_y<0||(best_y==start_y&&best_x==start_x)) continue;
        if (try_dfs_from(ai,best_y,best_x,obstacles,grid_size,free_cells)) return;
    }

    /* 阶段3：BFS 分层序退化 */
    build_bfs_order(ai,obstacles,grid_size);
}

/* ==================== 基础辅助 ==================== */
static inline bool ai_is_valid(SnakeAI *ai, int y, int x,
                                bool obstacles[GS][GS], BodyList *body) {
    if (y<0||y>=ai->grid_size||x<0||x>=ai->grid_size) return false;
    if (obstacles[y][x]) return false;
    return !body_contains(body,(Pos){y,x});
}

static int ai_count_solids(SnakeAI *ai, int y, int x, bool obstacles[GS][GS]) {
    int count=0;
    for (int d=0;d<4;d++) {
        int ny=y+DY[d],nx=x+DX[d];
        if (ny<0||ny>=ai->grid_size||nx<0||nx>=ai->grid_size) count++;
        else if (obstacles[ny][nx]) count++;
        else if (g_body_grid[ny][nx]) count++;
    }
    return count;
}

static int ai_flood_fill(SnakeAI *ai, Pos start,
                          bool obstacles[GS][GS], BodyList *body) {
    bool visited[GS][GS]; memset(visited,0,sizeof(visited));
    for (int i=0;i<body->len;i++) {
        Pos p=body_get(body,i); visited[p.y][p.x]=true;
    }
    for (int y=0;y<ai->grid_size;y++)
        for (int x=0;x<ai->grid_size;x++)
            if (obstacles[y][x]) visited[y][x]=true;
    int count=0;
    Pos queue[BFS_QUEUE_SIZE]; int qh=0,qt=0;
    for (int d=0;d<4;d++) {
        int ny=start.y+DY[d],nx=start.x+DX[d];
        if (ny>=0&&ny<ai->grid_size&&nx>=0&&nx<ai->grid_size&&!visited[ny][nx]) {
            visited[ny][nx]=true; queue[qt++]=(Pos){ny,nx}; count++;
        }
    }
    while (qh<qt) {
        Pos cur=queue[qh++];
        for (int d=0;d<4;d++) {
            int ny=cur.y+DY[d],nx=cur.x+DX[d];
            if (ny>=0&&ny<ai->grid_size&&nx>=0&&nx<ai->grid_size&&!visited[ny][nx]) {
                visited[ny][nx]=true; queue[qt++]=(Pos){ny,nx}; count++;
            }
        }
    }
    return count;
}

static bool ai_is_tail_reachable(SnakeAI *ai, Pos head,
                                   bool obstacles[GS][GS], BodyList *body) {
    if (body->len<=2) return true;
    Pos tail=body_tail(body);
    bool visited[GS][GS]; memset(visited,0,sizeof(visited));
    for (int i=0;i<body->len-1;i++) {
        Pos p=body_get(body,i); visited[p.y][p.x]=true;
    }
    for (int y=0;y<ai->grid_size;y++)
        for (int x=0;x<ai->grid_size;x++)
            if (obstacles[y][x]) visited[y][x]=true;
    visited[head.y][head.x]=true;
    Pos queue[BFS_QUEUE_SIZE]; int qh=0,qt=0;
    queue[qt++]=head;
    while (qh<qt) {
        Pos cur=queue[qh++];
        if (cur.y==tail.y&&cur.x==tail.x) return true;
        for (int d=0;d<4;d++) {
            int ny=cur.y+DY[d],nx=cur.x+DX[d];
            if (ny>=0&&ny<ai->grid_size&&nx>=0&&nx<ai->grid_size) {
                if ((ny==tail.y&&nx==tail.x)||!visited[ny][nx]) {
                    visited[ny][nx]=true; queue[qt++]=(Pos){ny,nx};
                }
            }
        }
    }
    return false;
}

static void ai_simulate_one_step(BodyList *out_body, int *out_steps, int *out_buf,
                                  int d, BodyList *body, int steps, int growth_buffer,
                                  int n_steps_grow, Pos food) {
    body_copy(out_body,body);
    int st=steps+1, buf=growth_buffer;
    Pos head=body_head(body);
    Pos nh={head.y+DY[d],head.x+DX[d]};
    bool eat=(nh.y==food.y&&nh.x==food.x);
    bool grow=(st%n_steps_grow==0);
    if (eat||grow) buf++;
    if (buf>0) buf--;
    else body_pop_back(out_body);
    body_push_front(out_body,nh);
    *out_steps=st; *out_buf=buf;
}

/* ==================== A* ==================== */
typedef struct { int f,g; Pos pos; } AStarNode;
typedef struct { AStarNode data[MAX_HEAP]; int size; } MinHeap;

static void heap_push(MinHeap *h, AStarNode node) {
    int i=h->size++; h->data[i]=node;
    while (i>0) {
        int p=(i-1)/2;
        if (h->data[p].f<=h->data[i].f) break;
        AStarNode tmp=h->data[p]; h->data[p]=h->data[i]; h->data[i]=tmp; i=p;
    }
}
static AStarNode heap_pop(MinHeap *h) {
    AStarNode top=h->data[0]; h->data[0]=h->data[--h->size];
    int i=0;
    while (1) {
        int l=2*i+1,r=2*i+2,s=i;
        if (l<h->size&&h->data[l].f<h->data[s].f) s=l;
        if (r<h->size&&h->data[r].f<h->data[s].f) s=r;
        if (s==i) break;
        AStarNode tmp=h->data[s]; h->data[s]=h->data[i]; h->data[i]=tmp; i=s;
    }
    return top;
}

static int ai_edge_hugging_a_star(SnakeAI *ai, Pos start, Pos target,
                                   bool obstacles[GS][GS], BodyList *body,
                                   int n_steps_grow, int out_path[MAX_ASTAR_PATH]) {
    bool static_body[GS][GS]; memset(static_body,0,sizeof(static_body));
    for (int i=0;i<body->len;i++) {
        Pos p=body_get(body,i); static_body[p.y][p.x]=true;
    }
    int best_g[GS][GS]; memset(best_g,0x7F,sizeof(best_g));
    best_g[start.y][start.x]=0;
    Pos par[GS][GS]; int pd[GS][GS]; memset(pd,-1,sizeof(pd));
    static MinHeap heap; heap.size=0;
    heap_push(&heap,(AStarNode){0,0,start});
    while (heap.size>0) {
        AStarNode cur=heap_pop(&heap);
        if (cur.pos.y==target.y&&cur.pos.x==target.x) {
            int tmp[MAX_ASTAR_PATH],len=0; Pos p=target;
            while (!(p.y==start.y&&p.x==start.x)) { tmp[len++]=pd[p.y][p.x]; p=par[p.y][p.x]; }
            for (int i=0;i<len;i++) out_path[i]=tmp[len-1-i];
            return len;
        }
        if (cur.g>best_g[cur.pos.y][cur.pos.x]) continue;
        for (int d=0;d<4;d++) {
            int ny=cur.pos.y+DY[d],nx=cur.pos.x+DX[d];
            if (ny<0||ny>=ai->grid_size||nx<0||nx>=ai->grid_size) continue;
            if (static_body[ny][nx]||obstacles[ny][nx]) continue;
            int solids=0;
            for (int id=0;id<4;id++) {
                int n2y=ny+DY[id],n2x=nx+DX[id];
                if (n2y<0||n2y>=ai->grid_size||n2x<0||n2x>=ai->grid_size) solids++;
                else if (obstacles[n2y][n2x]) solids++;
                else if (static_body[n2y][n2x]) solids++;
            }
            int pw=(n_steps_grow<=4)?10:(n_steps_grow<=16)?6:(n_steps_grow<=64)?3:1;
            int new_g=cur.g+1+pw*(4-solids);
            if (new_g<best_g[ny][nx]) {
                best_g[ny][nx]=new_g; par[ny][nx]=cur.pos; pd[ny][nx]=d;
                int ady=ny-target.y,adx=nx-target.x;
                if(ady<0)ady=-ady; if(adx<0)adx=-adx;
                heap_push(&heap,(AStarNode){new_g+ady+adx,new_g,(Pos){ny,nx}});
            }
        }
    }
    return 0;
}

/* ==================== BFS ==================== */
static int ai_bfs_path_to(SnakeAI *ai, Pos start, Pos target,
                            bool obstacles[GS][GS], BodyList *body,
                            int out_path[MAX_ASTAR_PATH]) {
    bool blocked[GS][GS]; memset(blocked,0,sizeof(blocked));
    for (int i=0;i<body->len;i++) {
        Pos p=body_get(body,i); blocked[p.y][p.x]=true;
    }
    for (int y=0;y<ai->grid_size;y++)
        for (int x=0;x<ai->grid_size;x++)
            if (obstacles[y][x]) blocked[y][x]=true;
    blocked[target.y][target.x]=false;
    bool visited[GS][GS]; memset(visited,0,sizeof(visited));
    visited[start.y][start.x]=true;
    Pos par[GS][GS]; int pd[GS][GS]; memset(pd,-1,sizeof(pd));
    Pos queue[BFS_QUEUE_SIZE]; int qh=0,qt=0;
    queue[qt++]=start;
    while (qh<qt) {
        Pos cur=queue[qh++];
        if (cur.y==target.y&&cur.x==target.x) {
            int tmp[MAX_ASTAR_PATH],len=0; Pos p=target;
            while (!(p.y==start.y&&p.x==start.x)) { tmp[len++]=pd[p.y][p.x]; p=par[p.y][p.x]; }
            for (int i=0;i<len;i++) out_path[i]=tmp[len-1-i];
            return len;
        }
        for (int d=0;d<4;d++) {
            int ny=cur.y+DY[d],nx=cur.x+DX[d];
            if (ny>=0&&ny<ai->grid_size&&nx>=0&&nx<ai->grid_size
                &&!visited[ny][nx]&&!blocked[ny][nx]) {
                visited[ny][nx]=true; par[ny][nx]=cur; pd[ny][nx]=d;
                queue[qt++]=(Pos){ny,nx};
            }
        }
    }
    return 0;
}

/* ==================== 哈密顿序工具 ==================== */
static inline int ham_offset(SnakeAI *ai, int from_idx, int to_idx) {
    if (to_idx>=from_idx) return to_idx-from_idx;
    return ai->hamilton_len-from_idx+to_idx;
}

/* 在哈密顿序上向前扫描，找第一个与 head 相邻且在 safe_moves 中的方向 */
static int hamilton_next_safe(SnakeAI *ai, Pos head,
                               int safe_moves[], int num_safe, int scan_limit) {
    int hidx=ai->hamilton_pos[head.y][head.x];
    if (hidx<0||ai->hamilton_len==0) return -1;
    for (int skip=1;skip<=scan_limit;skip++) {
        int scan_idx=(hidx+skip)%ai->hamilton_len;
        Pos sp=ai->hamilton_cycle[scan_idx];
        int dy=sp.y-head.y, dx=sp.x-head.x;
        if (abs(dy)+abs(dx)!=1) continue;
        for (int d=0;d<4;d++) {
            if (DY[d]==dy&&DX[d]==dx) {
                for (int i=0;i<num_safe;i++)
                    if (safe_moves[i]==d) return d;
                break;
            }
        }
    }
    return -1;
}

/* ==================== A* 路径安全验证 ==================== */
static bool ai_verify_astar_path(SnakeAI *ai, SnakeEnv *e,
                                  int path[], int path_len) {
    BodyList sim_body; body_copy(&sim_body,&e->body);
    int sim_steps=e->steps, sim_buf=e->growth_buffer;
    Pos sim_food=e->food;
    for (int i=0;i<path_len;i++) {
        int d=path[i];
        Pos h=body_head(&sim_body);
        Pos nh={h.y+DY[d],h.x+DX[d]};
        bool eat=(nh.y==sim_food.y&&nh.x==sim_food.x);
        bool grow=((sim_steps+1)%e->n_steps_grow==0);
        if (eat||grow) sim_buf++;
        if (eat) {
            if (sim_buf>0) sim_buf--; else body_pop_back(&sim_body);
            body_push_front(&sim_body,nh); sim_steps++;
            return ai_is_tail_reachable(ai,nh,e->obstacles,&sim_body);
        }
        if (sim_buf>0) sim_buf--; else body_pop_back(&sim_body);
        body_push_front(&sim_body,nh); sim_steps++;
    }
    return ai_is_tail_reachable(ai,body_head(&sim_body),e->obstacles,&sim_body);
}

/* ============================================================
 * 智能开阔地 fallback（替代原来的单步 solid 贪心）
 *
 * 对每个 safe_moves[i] 计算综合评分：
 *   score = w1*reachable + w2*tail_bonus + w3*ham_progress + w4*food_bonus
 *
 *   reachable    : flood_fill 可达空间（主项，防死胡同）
 *   tail_bonus   : 尾部可达则加分（保护性）
 *   ham_progress : 若 hamilton_valid，走向哈密顿序更大偏移的方向加分
 *                  若 !hamilton_valid，走向哈密顿序编号较大（BFS分层靠后）的方向加分
 *   food_bonus   : 曼哈顿距离食物更近加分（机会性）
 *
 * 权重根据 N 动态调整：
 *   N>=256 重保护性（reachable + tail），弱化 food
 *   N<256  适当兼顾 food
 * ============================================================ */
static int ai_smart_fallback(SnakeAI *ai, SnakeEnv *e,
                              int safe_moves[], int safe_move_reach[],
                              BodyList safe_move_simbody[], int num_safe) {
    int n = e->n_steps_grow;
    Pos head = body_head(&e->body);

    /* 权重配置 */
    int w_reach, w_tail, w_ham, w_food;
    if (n >= 256) {
        w_reach = 1000; w_tail = 800; w_ham = 60; w_food = 5;
    } else if (n >= 64) {
        w_reach = 800;  w_tail = 500; w_ham = 80; w_food = 20;
    } else if (n >= 16) {
        w_reach = 600;  w_tail = 300; w_ham = 60; w_food = 40;
    } else {
        w_reach = 400;  w_tail = 150; w_ham = 30; w_food = 80;
    }

    int head_hidx = (ai->hamilton_len > 0) ? ai->hamilton_pos[head.y][head.x] : -1;
    int best_d = safe_moves[0];
    long long best_score = -1;

    for (int i = 0; i < num_safe; i++) {
        int d = safe_moves[i];
        BodyList *sb = &safe_move_simbody[i];
        Pos sh = body_head(sb);

        /* 1. 可达空间（已算好） */
        long long score = (long long)w_reach * safe_move_reach[i];

        /* 2. 尾部可达加分 */
        bool tail_ok = ai_is_tail_reachable(ai, sh, e->obstacles, sb);
        if (tail_ok) score += w_tail * 100;

        /* 3. 哈密顿序进度：走向哈密顿序编号更靠前（正向进度更大）的方向加分 */
        if (ai->hamilton_len > 0) {
            int sh_hidx = ai->hamilton_pos[sh.y][sh.x];
            if (head_hidx >= 0 && sh_hidx >= 0) {
                /* 正向偏移：sh 在 head 哈密顿序的前方距离 */
                int progress = ham_offset(ai, head_hidx, sh_hidx);
                /* progress==1 最好（正好下一个），越大越绕 */
                int ham_sc = (progress == 1) ? 100 :
                             (progress <= 3) ? 60  :
                             (progress <= ai->grid_size) ? 30 : 0;
                score += (long long)w_ham * ham_sc;
            }
        }

        /* 4. 食物距离（靠近食物加分，但权重低） */
        int fdy = sh.y - e->food.y, fdx = sh.x - e->food.x;
        if (fdy<0) fdy=-fdy; if (fdx<0) fdx=-fdx;
        int food_dist = fdy + fdx;
        /* 距离越近分越高，最多加 200 分 */
        int max_dist = ai->grid_size * 2;
        int food_sc = (max_dist - food_dist) * 100 / max_dist;
        score += (long long)w_food * food_sc;

        /* 5. 当前格邻居 solid 数惩罚（越贴墙越差，但权重很低） */
        int ny = head.y + DY[d], nx = head.x + DX[d];
        int solids = ai_count_solids(ai, ny, nx, e->obstacles);
        score -= (long long)10 * solids;

        if (score > best_score) {
            best_score = score;
            best_d = d;
        }
    }
    return best_d;
}

/* ==================== get_action ==================== */
static int ai_get_action(SnakeAI *ai, SnakeEnv *e) {
    Pos head = body_head(&e->body);
    int n    = e->n_steps_grow;
    body_grid_build(&e->body);

    /* ---- 安全扫描 ---- */
    int safe_moves[4], safe_move_reach[4];
    BodyList safe_move_simbody[4];
    int num_safe=0, best_unsafe_move=0, max_unsafe_area=-1;

    for (int d=0;d<4;d++) {
        int ny=head.y+DY[d],nx=head.x+DX[d];
        bool will_grow=(e->growth_buffer>0)||(ny==e->food.y&&nx==e->food.x)
                       ||((e->steps+1)%n==0);
        BodyList check_body; body_copy(&check_body,&e->body);
        if (!will_grow) body_pop_back(&check_body);
        if (!ai_is_valid(ai,ny,nx,e->obstacles,&check_body)) continue;

        BodyList sim_body; int sim_steps,sim_buf;
        ai_simulate_one_step(&sim_body,&sim_steps,&sim_buf,
                             d,&e->body,e->steps,e->growth_buffer,n,e->food);
        Pos sim_head=body_head(&sim_body);
        int reachable=ai_flood_fill(ai,sim_head,e->obstacles,&sim_body);
        int expected=e->grid_size*e->grid_size-e->num_obstacles-sim_body.len;

        bool tail_ok =ai_is_tail_reachable(ai,sim_head,e->obstacles,&sim_body);
        bool space_ok=(reachable>=sim_body.len);
        bool is_safe =(reachable==expected)||tail_ok||space_ok;

        if (is_safe) {
            safe_moves[num_safe]=d;
            safe_move_reach[num_safe]=reachable;
            body_copy(&safe_move_simbody[num_safe],&sim_body);
            num_safe++;
        } else {
            if (reachable>max_unsafe_area) { max_unsafe_area=reachable; best_unsafe_move=d; }
        }
    }

    if (num_safe==0) return best_unsafe_move;

    /* ---- N=1 策略 ---- */
    if (n==1) {
        int astar_path[MAX_ASTAR_PATH];
        int astar_len=ai_edge_hugging_a_star(ai,head,e->food,e->obstacles,&e->body,n,astar_path);
        if (astar_len>0&&ai_verify_astar_path(ai,e,astar_path,astar_len)) {
            int fd=astar_path[0];
            for (int i=0;i<num_safe;i++) if (safe_moves[i]==fd) return fd;
        }
        int bfs_path[MAX_ASTAR_PATH];
        int bfs_len=ai_bfs_path_to(ai,head,e->food,e->obstacles,&e->body,bfs_path);
        if (bfs_len>0) {
            int fd=bfs_path[0]; for (int i=0;i<num_safe;i++) if(safe_moves[i]==fd) return fd;
        }
        return ai_smart_fallback(ai,e,safe_moves,safe_move_reach,safe_move_simbody,num_safe);
    }

    /* ---- N>1 策略 ---- */

    /* 策略1：A* + 尾部验证 */
    int astar_path[MAX_ASTAR_PATH];
    int astar_len=ai_edge_hugging_a_star(ai,head,e->food,e->obstacles,&e->body,n,astar_path);
    if (astar_len>0&&ai_verify_astar_path(ai,e,astar_path,astar_len)) {
        int fd=astar_path[0];
        for (int i=0;i<num_safe;i++) if (safe_moves[i]==fd) return fd;
    }

    /* 策略2：BFS 备选（小 N） */
    if (n<=16) {
        int bfs_path[MAX_ASTAR_PATH];
        int bfs_len=ai_bfs_path_to(ai,head,e->food,e->obstacles,&e->body,bfs_path);
        if (bfs_len>0) {
            int fd=bfs_path[0]; for (int i=0;i<num_safe;i++) if(safe_moves[i]==fd) return fd;
        }
    }

    /* 策略3：哈密顿跟随（N>=32） */
    if (n>=32&&ai->hamilton_len>0) {
        /* 3a：捷径优化（N>=64，食物在哈密顿前方半程内且安全） */
        if (n>=64) {
            int hidx=ai->hamilton_pos[head.y][head.x];
            int fidx=ai->hamilton_pos[e->food.y][e->food.x];
            if (hidx>=0&&fidx>=0) {
                int food_off=ham_offset(ai,hidx,fidx);
                if (food_off>0&&food_off<=ai->hamilton_len/2) {
                    int bfs_path[MAX_ASTAR_PATH];
                    int bfs_len=ai_bfs_path_to(ai,head,e->food,e->obstacles,&e->body,bfs_path);
                    if (bfs_len>0&&ai_verify_astar_path(ai,e,bfs_path,bfs_len)) {
                        int fd=bfs_path[0];
                        for (int i=0;i<num_safe;i++) if(safe_moves[i]==fd) return fd;
                    }
                }
            }
        }

        /* 3b：哈密顿跟随（真回路 scan=4，退化序 scan=grid_size*2） */
        int scan_limit=ai->hamilton_valid ? 4 : ai->grid_size*2;
        int ham_d=hamilton_next_safe(ai,head,safe_moves,num_safe,scan_limit);
        if (ham_d>=0) return ham_d;
    }

    /* 策略4：智能开阔地 fallback（替代原来的单步 solid 贪心） */
    return ai_smart_fallback(ai,e,safe_moves,safe_move_reach,safe_move_simbody,num_safe);
}

/* ==================== 运行游戏 ==================== */
static int run_game(int n_steps, bool noout) {
    SnakeEnv env;
    env.grid_size=GS; env.n_steps_grow=n_steps; env.num_obstacles=NUM_OBSTACLES;
    env_reset(&env); body_grid_build(&env.body);
    SnakeAI ai; ai_init(&ai,GS,env.obstacles);
    while (1) {
        int action=ai_get_action(&ai,&env);
        int result=env_step(&env,action);
        if (!noout) {
            char grid[GS][GS]; memset(grid,'.',sizeof(grid));
            for (int y=0;y<GS;y++)
                for (int x=0;x<GS;x++)
                    if (env.obstacles[y][x]) grid[y][x]='X';
            for (int i=0;i<env.body.len;i++) {
                Pos p=body_get(&env.body,i); grid[p.y][p.x]='O';
            }
            Pos hd=body_head(&env.body);
            grid[hd.y][hd.x]='@'; grid[env.food.y][env.food.x]='*';
            printf("\033c");
            for (int y=0;y<GS;y++) {
                for (int x=0;x<GS;x++) printf("%c ",grid[y][x]);
                printf("\n");
            }
            printf("规则: 每 %d 步生长 | 步数: %d | 果实: %d | 身体: %d\n",
                   env.n_steps_grow,env.steps,env.score,env.body.len);
        }
        if (result!=0) {
            if (!noout) printf("游戏结束！原因: %s | 最终吃到果实: %d\n",
                               result==1?"撞墙":result==2?"撞障碍物":"咬到自己",env.score);
            return env.score;
        }
    }
}

/* ==================== 批量测试 ==================== */
static void batch_test(int n_steps, int trials) {
    int total=0, scores[200];
    rng_seed(42);
    unsigned long long seeds[200];
    for (int i=0;i<200;i++) seeds[i]=rng_next();
    for (int i=0;i<trials;i++) {
        rng_seed(seeds[i]); scores[i]=run_game(n_steps,true); total+=scores[i];
    }
    printf("N=%d trials=%d avg=%.2f total=%d\n[",n_steps,trials,(double)total/trials,total);
    for (int i=0;i<trials;i++) { printf("%d",scores[i]); if(i<trials-1) printf(", "); }
    printf("]\n");
}

int main0(int argc,char *argv[]) {
    int n_steps=1; bool noout=false;
    if (argc>=2) n_steps=atoi(argv[1]);
    if (argc>=3) noout=(atoi(argv[2])!=0);
    if (n_steps==0) {
        int test_ns[]={1,2,4,8,16,32,64,128,256,512};
        for (int i=0;i<10;i++) batch_test(test_ns[i],20);
        return 0;
    }
    rng_seed((unsigned long long)time(NULL));
    int score=run_game(n_steps,noout);
    if (noout) printf("%d\n",score);
    return 0;
}

/* ==================== 打印最终地图 ==================== */
void print_final_map(const SnakeEnv *env) {
    char out[20][25];
    for (int i=0;i<20;i++) strcpy(out[i],raw_map[i]);
    for (int i=0;i<20;i++) for (int j=1;j<19;j++) out[i][j]='.';
    for (int i=0;i<20;i++) {
        out[0][i]='#'; out[19][i]='#'; out[i][0]='#'; out[i][19]='#';
    }
    for (int y=0;y<GS;y++)
        for (int x=0;x<GS;x++)
            if (env->obstacles[y][x]) out[y+1][x+1]='O';
    for (int i=0;i<env->body.len;i++) {
        Pos p=body_get(&env->body,i); out[p.y+1][p.x+1]='B';
    }
    Pos hd=body_head(&env->body);
    out[hd.y+1][hd.x+1]='H';
    if (env->food.y<GS&&env->food.y>=0&&env->food.x<GS&&env->food.x>=0)
        out[env->food.y+1][env->food.x+1]='F';
    for (int y=0;y<GS+2;y++) {
        for (int x=0;x<GS+2;x++) printf("%c",out[y][x]);
        printf("\n");
    }
    printf("%d",env->score);
}

/* ==================== main（OJ 交互） ==================== */
int main() {
    setvbuf(stdout,NULL,_IOFBF,4096);
    setvbuf(stdin, NULL,_IOFBF,4096);
    for (int i=0;i<20;i++) scanf("%s",raw_map[i]);
    int N; scanf("%d",&N);
    rng_seed((unsigned long long)time(NULL));

    SnakeEnv env;
    env.grid_size=GS; env.n_steps_grow=N;
    env.num_obstacles=0; env_reset(&env); env.num_obstacles=NUM_OBSTACLES;

    int head_y=-1,head_x=-1;
    for (int i=1;i<=18;i++) {
        for (int j=1;j<=18;j++) {
            char c=raw_map[i][j]; int ey=i-1,ex=j-1;
            if (c=='H') { head_y=ey; head_x=ex; }
            else if (c=='O'||c=='#') env.obstacles[ey][ex]=true;
            else if (c=='F') env.food=(Pos){ey,ex};
        }
    }
    body_init(&env.body,(Pos){head_y,head_x});
    bool body_visited[18][18]={false};
    body_visited[head_y][head_x]=true;
    while (true) {
        bool found_next=false;
        for (int d=0;d<4;d++) {
            Pos tail=body_tail(&env.body);
            int ny=tail.y+DY[d],nx=tail.x+DX[d];
            if (ny>=0&&ny<18&&nx>=0&&nx<18&&!body_visited[ny][nx]) {
                if (raw_map[ny+1][nx+1]=='B'||raw_map[ny+1][nx+1]=='b') {
                    body_add_tail(&env.body,(Pos){ny,nx});
                    body_visited[ny][nx]=true; found_next=true; break;
                }
            }
        }
        if (!found_next) break;
    }
    body_grid_build(&env.body);
    SnakeAI ai; ai_init(&ai,GS,env.obstacles);

    while (1) {
        int action=ai_get_action(&ai,&env);
        printf("%c\n%d\n",DIR_CHARS[action],env.score);
        fflush(stdout);
        int res_y,res_x;
        if (scanf("%d %d",&res_y,&res_x)!=2) break;
        if (res_y==100&&res_x==100) { print_final_map(&env); break; }
        env_step(&env,action);
        if (res_y!=20&&res_x!=20) { env.food.y=res_y-1; env.food.x=res_x-1; }
    }
    return 0;
}
```
